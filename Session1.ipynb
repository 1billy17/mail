## Подключение пакетов

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
```

```python
file_name = 'datasert.xls'
output_name = 'result.xlsx'
```

```python
table = pd.read_excel(file_name)
```

```python
table.head(10)
```

```python
table = table.drop(['№ п/п','№\nдокумента','Исх. №\nДата', 'Примечание'], axis=1)
table.head(2)
```

```python
table[['автор','компания']] = table['Автор'].str.split('(',n=1,expand=True)
```

```python
table.head(2)
```

```python
table = table.drop(['Автор'], axis=1)
```

```python
table.head(2)
```

```python
table = table.rename(
    columns={
        'Дата\nрегистрации': 'Датарегистрации',
})
```

```python
table.head(2)
```

```python
table['Адресаты']='' 
```

```python
слова = ['Вордскиллс', 'соревнован', 'чемпион']

for слово in слова:
    table.loc[table['Краткое\nсодержание'].str.contains(слово, case=False), 'Адресат'] = 'Тымчикова А. Ю.'

table.head(10)
```

```python
table.to_excel(output_name, index=False)
```
